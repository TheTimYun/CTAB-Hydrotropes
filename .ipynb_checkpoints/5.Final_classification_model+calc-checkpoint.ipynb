```python
import numpy as np
import pandas as pd
from rdkit import RDConfig
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier, XGBRegressor
from rdkit import Chem
from molfeat.calc import RDKitDescriptors2D, RDKitDescriptors3D
from molfeat.trans import MoleculeTransformer
import datamol as dm
from rdkit.Chem import AllChem

from functools import partial
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score, precision_score, roc_curve
from sklearn.ensemble import RandomForestClassifier
```

First, we load set of descriptors, target sets and features to drop (we found them during initial descriptors calculation in the very first notebook)

```python
with open('features_to_drop.pickle', 'rb') as inp:
    features_to_drop = pickle.load(inp)
```

```python
with open('X_standard_dropped.pickle', 'rb') as inp:
    X_standard = pickle.load(inp)
```

```python
with open('y.pickle', 'rb') as inp:
    y = pickle.load(inp)
```

```python
with open('viscosity.pickle', 'rb') as inp:
    viscosity = pickle.load(inp)
viscosity = np.log(viscosity/1000)
```

**Classification**

The base of the model is XGBoost Classifier, which uses features, scaled by MinMaxScaler. Parameters of XGBoost were found earlier.   
We initialize the model and train it with X_standard dataset

```python
trained_model_clf = make_pipeline(MinMaxScaler(), XGBClassifier(n_jobs = -1, learning_rate = 0.09362664231796851,
                                                      max_depth = 83,
                                                      n_estimators = 22,
                                                      reg_alpha = 0.002211107301032561))
```

```python
trained_model_clf.fit(X_standard, y)
```

Let's save the model

```python
with open('trained_model_clf.pickle', 'wb') as output:
    pickle.dump(trained_model_clf, output)
```

And draw feature importances

```python
fig = plt.figure(figsize = (12,8))
clf = trained_model_clf.steps[1][1]
plt.barh(width = clf.feature_importances_[np.argsort(clf.feature_importances_)[-15:]], 
         y = X_standard.columns[np.argsort(clf.feature_importances_)[-15:]])
```

Its interesting, than only one system descriptor (CTAB concentration) appears in 15 most important features. Other features are RDKit descriptors. 
Most of them (RDF, MORSE, AUTOCORR3D) are 3D descriptors. Only one absolute descriptor (MinMaxPartialCharge) and only one fingerprint are among the strongest. It seems, that 3D structure of additive and its interaction with CTAB are the most important for gel formation. 


**Regression**

The base of the model is XGBoost Regressor, which uses features, scaled by MinMaxScaler. Parameters of XGBoost were found earlier.   
We initialize the model and train it with X_standard dataset. Then we again save the model and find 15 most important descriptors.

```python
trained_model_reg = make_pipeline(MinMaxScaler(), XGBRegressor(grow_policy = 'depthwise', 
                                                  learning_rate = 1.1849939625930745,
                                                  max_depth = 10,
                                                  n_estimators = 26, 
                                                  reg_alpha = 0.0022376683053165628))

                                                  
```

```python
trained_model_reg.fit(X_standard, viscosity)
```

```python
with open('trained_model_reg.pickle', 'wb') as output:
    pickle.dump(trained_model_reg, output)
```

```python
fig = plt.figure(figsize = (12,8))
clf = trained_model_reg.steps[1][1]
plt.barh(width = clf.feature_importances_[np.argsort(clf.feature_importances_)[-15:]], 
         y = X_standard.columns[np.argsort(clf.feature_importances_)[-15:]])
```

It's interesting to note, that the most important descriptor (with a huge gap between it and the nearest neighbor) is PEOE_VSA, which is actually descriptor of surface area, which takes into account gasteiger charges. Only several descriptors from previous step, which are CalcMORSE and Autocorrelation appear here. We can conclude, that the "strengh" of intermolecular interactions, which are responsible for the absolute value of viscosity, is depependent on other characterstics, than the "number" of interactions, that influence gel formation. 

**Сalculator**

Here you should hust follow what is written in input strings! Don't forget to activate embed_optmize function

```python
def embed_optimize(smi):
    m = Chem.MolFromSmiles(smi)
    m = Chem.AddHs(m)
    AllChem.EmbedMolecule(m)
    AllChem.MMFFOptimizeMolecule(m, maxIters=500)
    return m
```

```python
CTAB_conc = int(input('Enter CTAB Concnetration in mM '))
```

```python
additive_conentrations = input('Enter additive concentraions in mM like 50,80,60,90 ')
additive_conentrations = [int(x) for x in additive_conentrations.split(',')]
```

```python
temp = int(input('Enter the temperature in Celcius degrees '))
```

```python
SMILES = input('Enter the SMILES of the additive ')
add = embed_optimize(SMILES)
```

Here is the function to create dataframe for models. The steps are the same, as in "*1.Hydrotropes.ipynb*" notebook

```python
#fingerprints calculation
fp = AllChem.GetMorganFingerprintAsBitVect(add, radius = 3)
fp_df = pd.DataFrame(np.array(fp).reshape(1,-1), columns = ['fp{}'.format(i) for i in range(2048)])
#descriptors calculation
calc_2D = RDKitDescriptors2D()
calc_3D = RDKitDescriptors3D()
with dm.without_rdkit_log():
    feats_2D = calc_2D(add)
    feats_3D = calc_3D(add)
#create one big features df
features = pd.DataFrame(np.concatenate([feats_2D, feats_3D]).reshape(1,-1), columns=calc_2D.columns + calc_3D.columns)
features = pd.concat([fp_df, features], axis = 1)
#create the df with characteritics of the system (concentration and temperature)
dic = {'CTAB concentration (mM)':[], 'Additive concentration':[], 'CTAB/additive':[], 'Temperature':[] }
for conc in additive_conentrations:
    dic['CTAB concentration (mM)'].append(CTAB_conc)
    dic['Additive concentration'].append(conc)
    dic['CTAB/additive'].append(CTAB_conc/conc)
    dic['Temperature'].append(temp)
df = pd.DataFrame(dic)
#create X_dataframe with several instances - different concentrations of the same additive
X = pd.concat([df, pd.concat([features]*len(additive_conentrations), ignore_index= True)], axis = 1)
#drop unneseccary features
X.drop(columns=['Alerts'], inplace=True)
X = X.drop(columns = features_to_drop)
```

Here you should just activate two cells and get the result. 1 is gel, 0 is non-gel, viscosity is given in cP

```python
classes = trained_model_clf.predict(X)
viscosity = 1000*np.exp(trained_model_reg.predict(X))
```
